```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
#notebook mode
%matplotlib inline  
```

```python
df = pd.read_csv("Clean/reports21data.csv",sep = ";",decimal = ".",encoding = "latin-1")
```

```python
def trace(x,tl,labelX= None,labelY= None) :
    fig = plt.figure(figsize=(12, 6))
    sqft = fig.add_subplot(121)
    cost = fig.add_subplot(122)

    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    labels = x.index
    sizes = x
    explode = tuple([0.1*i for i in range(len(x))])
    sqft.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    sqft.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    sqft.set_title(tl)
    
    #hist chart
    plt.rcdefaults()
    y_pos = np.arange(len(labels))
    error = np.random.rand(len(labels))

    cost.barh(y_pos, sizes, xerr=error, align='center', color='blue', ecolor='black')
    cost.set_yticks(y_pos)
    cost.set_yticklabels(labels)
    cost.invert_yaxis()  # labels read top-to-bottom
    cost.set_title(tl)

    plt.show()
        
```

```python
name_list=['LEFT CURRENT [A]', 'LEFT POWER [W]',
       'LEFT RESISTANCE [Ohm]', 'LEFT VOLTAGE [V]', 'RIGHT CURRENT [A]',
       'RIGHT POWER [W]', 'RIGHTRESISTANCE [Ohm]', 'RIGHT VOLTAGE [V]',
       'WIRE FUNCTIONAL', 'WIRE  DISTANCE', 'TWISTED WIRES',
       'THERMAL INSPECTION', 'BUSBAR POSITION', 'BUSBAR DIAMETER',
       'CYCLE TIME[Sn]']
```

```python
for i in name_list:
    x = df[i].value_counts()
    tl=i
    trace(x,tl)
```
